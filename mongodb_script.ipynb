```python
import pandas as pd
from pymongo import MongoClient

# Load the CSV file into a pandas DataFrame
data_path = "/home/mutayyab/Documents/dataset/preprocessed_audios.csv"  # Update with the correct path
df = pd.read_csv(data_path)

# Display the first few rows of the DataFrame to ensure data is loaded correctly
print("First few rows of the DataFrame:")
print(df.head())

# MongoDB connection details
mongo_uri = "mongodb://localhost:27017/"  # Update with your MongoDB URI
db_name = "Preprocessed_audios"  # Update with your MongoDB database name
collection_name = "Features"  # Update with your MongoDB collection name

# Connect to MongoDB
print("\nConnecting to MongoDB...")
client = MongoClient(mongo_uri)
db = client[db_name]
collection = db[collection_name]

# Check if the collection already exists
if collection_name in db.list_collection_names():
    print(f"\nCollection '{collection_name}' already exists. Dropping the existing collection...")
    db.drop_collection(collection_name)

# Convert DataFrame records to dictionary format and insert into MongoDB
print("\nInserting records into MongoDB collection...")
records = df.to_dict(orient='records')
result = collection.insert_many(records)

# Print the result of the insert operation
print("\nInsert operation result:")
print(result)

# Count the number of documents inserted
print(f"\nNumber of documents inserted: {len(result.inserted_ids)}")

# Close MongoDB connection
print("\nClosing MongoDB connection...")
client.close()

print("\nData successfully stored in MongoDB.")
```
